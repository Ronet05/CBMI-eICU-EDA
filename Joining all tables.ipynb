<h3>Now we will merge everything into one table</h3>

```python
#Change the name of timeoffset columns for different tables into a common one, here we will call it offset
labs=labs.rename(columns={'labresultoffset':'offset'})
vasopres=vasopres.rename(columns={'infusionoffset':'offset'})
gcs=gcs.rename(columns={'nursingchartentryoffset':'offset'})
```

```python
dfs = [labs,gcs,vasopres]
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on=['patientunitstayid','offset']), dfs)
```
